# Facebook Ads Weekly Report

## Import data and update some values manually

```python
import pandas as pd

# read file
df = pd.read_excel('fb_raw_data.xlsx')

# Need to be updated
Days_in_lastmonth =  31
Days_analysed = 7
Thismonth = 'Sep 1-7'
Lastmonth = 'Aug Prorated'

# First column name in the Ads analysis
first_column_name = Thismonth + ' Ads'

# Select SG Campaign, Need to be updated if new campaign is created
values = ['C144_SG_2209_MLP_Cov_Purchase-Prop',
          'C145_SG_2209_MLP_Cov_Purchase-Gen',
          'C149_SG_2210_MLP_Cov_Purchase-Prop-SPG',
          'C159_SG_2303_MLP_Cov_Purchase-Retargeting',
         'C164_SG_2307_MLP_Cov_Purchase-MFF (Sub)',
          'C164_SG_2307_MLP_Cov_Purchase-MFF (Purch)',
         'C165_SG_2307_MLP_Cov_Purchase-Prop-APPSPG'] 
df_SG = df.loc[df['Campaign'].isin(values)]
# Simplify Campaign name, Need to be updated if new campaign is created
dic = {'C144_SG_2209_MLP_Cov_Purchase-Prop': 'APP',
    'C145_SG_2209_MLP_Cov_Purchase-Gen': 'Gen',
    'C149_SG_2210_MLP_Cov_Purchase-Prop-SPG': 'SPG',
    'C159_SG_2303_MLP_Cov_Purchase-Retargeting': 'Retargeting',
     'C164_SG_2307_MLP_Cov_Purchase-MFF (Sub)':'MFF',
       'C164_SG_2307_MLP_Cov_Purchase-MFF (Purch)':'MFF',
      'C165_SG_2307_MLP_Cov_Purchase-Prop-APPSPG':'Prop'} 
df_SG = df_SG.replace({'Campaign': dic})
# Filter data into different Campaigns, Need to be updated if new campaign is created
Retargeting_SG = df_SG.loc[df_SG['Campaign'] == 'Retargeting']
APP_SG = df_SG.loc[df_SG['Campaign'] == 'APP']
SPG_SG = df_SG.loc[df_SG['Campaign'] == 'SPG']
Gen_SG = df_SG.loc[df_SG['Campaign'] == 'Gen']
MFF_SG = df_SG.loc[df_SG['Campaign'] == 'MFF']
Prop_SG = df_SG.loc[df_SG['Campaign'] == 'Prop']
 
# Select UK Campaign, Need to be updated if new campaign is created
values = ['C157_UK_2302_MLP_Cov_Purchase-Prop', 'C162_UK_2305_MLP_Cov_Purchase-Retargeting', 'C163_UK_2305_MLP_Cov_Purchase-LimitPrice'] 
df_UK = df.loc[df['Campaign'].isin(values)]
# Simplify Campaign name, Need to be updated if new campaign is created
dic = {'C157_UK_2302_MLP_Cov_Purchase-Prop':'Prop',
'C162_UK_2305_MLP_Cov_Purchase-Retargeting':'Retargeting',
'C163_UK_2305_MLP_Cov_Purchase-LimitPrice':'Limitprice'} 
df_UK = df_UK.replace({'Campaign': dic})
# Filter data into different Campaigns, Need to be updated if new campaign is created
Retargeting_UK = df_UK.loc[df_UK['Campaign'] == 'Retargeting']
Prop_UK = df_UK.loc[df_UK['Campaign'] == 'Prop']
Limitprice_UK = df_UK.loc[df_UK['Campaign'] == 'Limitprice']


# Campaign orders, Need to be updated if new campaign is created
index_order = ['Retargeting','APP','SPG','Gen','MFF','Prop']

# Columns orders
columns_order_thismonth = ['Amount_spent', 'Impressions', 'Link_clicks', 'Adds_to_cart','Purchases', 'Revenue', 'ROAS', 'CPA', 'CR(%)']
columns_order_lastmonth = ['Amount_spent_lastmonth', 'Impressions_lastmonth', 'Link_clicks_lastmonth', 'Adds_to_cart_lastmonth','Purchases_lastmonth', 'Revenue_lastmonth', 'ROAS', 'CPA', 'CR(%)']
columns_order_ad = ['Amount_spent', 'Impressions', 'Link_clicks', 'Adds_to_cart','Purchases', 'Revenue', 'CTR(%)','ROAS', 'CPA', 'CR(%)']
```

## SG market

### Overall performance

```python
# Overall performance this month
Overall_SG_thismonth = df_SG[['Amount_spent', 'Impressions', 'Link_clicks', 'Adds_to_cart', 'Purchases', 'Revenue']].sum()
# Calculate the ROAS, CPA, and CR
Overall_SG_thismonth['ROAS'] = Overall_SG_thismonth['Revenue'] / Overall_SG_thismonth['Amount_spent']
Overall_SG_thismonth['CPA'] = Overall_SG_thismonth['Amount_spent'] / Overall_SG_thismonth['Purchases']
Overall_SG_thismonth['CR(%)'] = (Overall_SG_thismonth['Purchases'] / Overall_SG_thismonth['Link_clicks']) * 100


# Pro rata performance last month
Overall_SG_lastmonth = df_SG[['Amount_spent_lastmonth', 'Impressions_lastmonth', 'Link_clicks_lastmonth', 'Adds_to_cart_lastmonth', 'Purchases_lastmonth', 'Revenue_lastmonth']].sum()
# Calculate the pro rata
Overall_SG_lastmonth = Overall_SG_lastmonth / Days_in_lastmonth * Days_analysed
# Calculate the ROAS, CPA, and CR
Overall_SG_lastmonth['ROAS'] = Overall_SG_lastmonth['Revenue_lastmonth'] / Overall_SG_lastmonth['Amount_spent_lastmonth']
Overall_SG_lastmonth['CPA'] = Overall_SG_lastmonth['Amount_spent_lastmonth'] / Overall_SG_lastmonth['Purchases_lastmonth']
Overall_SG_lastmonth['CR(%)'] = (Overall_SG_lastmonth['Purchases_lastmonth'] / Overall_SG_lastmonth['Link_clicks_lastmonth']) * 100


# Combine two series
# Change the series into same index, in order to concat
Overall_SG_lastmonth.index = Overall_SG_thismonth.index
Overall_SG = pd.concat([Overall_SG_lastmonth,Overall_SG_thismonth],axis=1)
# Change index
Overall_SG = Overall_SG.reset_index()
# Change column name
Overall_SG.columns = ['Overall_SG', Lastmonth, Thismonth] 


# Calculate growth
Overall_SG['Growth(%)'] = (Overall_SG[Thismonth] - Overall_SG[Lastmonth]) / Overall_SG[Lastmonth] * 100
# Round 2
Overall_SG = Overall_SG.round(2)
```

### Campaign performance

```python
# SG Campaign performance this month
Campaign_SG_thismonth = pd.pivot_table(df_SG, index=['Campaign'], values=['Amount_spent', 'Impressions', 'Link_clicks', 'Adds_to_cart', 'Purchases', 'Revenue'], aggfunc='sum')
Campaign_SG_thismonth['ROAS'] = Campaign_SG_thismonth['Revenue'] / Campaign_SG_thismonth['Amount_spent']
Campaign_SG_thismonth['CPA'] = Campaign_SG_thismonth['Amount_spent'] / Campaign_SG_thismonth['Purchases']
Campaign_SG_thismonth['CR(%)'] = (Campaign_SG_thismonth['Purchases'] / Campaign_SG_thismonth['Link_clicks']) * 100
# Change order
Campaign_SG_thismonth = Campaign_SG_thismonth.reindex(columns=columns_order_thismonth)
Campaign_SG_thismonth = Campaign_SG_thismonth.reindex(index=index_order)
# Transpose
Campaign_SG_thismonth = Campaign_SG_thismonth.transpose()


# Pro rata performance last month
Campaign_SG_lastmonth = pd.pivot_table(df_SG, index=['Campaign'], values=['Amount_spent_lastmonth', 'Impressions_lastmonth', 'Link_clicks_lastmonth', 'Adds_to_cart_lastmonth', 'Purchases_lastmonth', 'Revenue_lastmonth'], aggfunc='sum')
# Calculate the pro rata
Campaign_SG_lastmonth = Campaign_SG_lastmonth / Days_in_lastmonth * Days_analysed
# Calculate the ROAS, CPA and CR
Campaign_SG_lastmonth['ROAS'] = Campaign_SG_lastmonth['Revenue_lastmonth'] / Campaign_SG_lastmonth['Amount_spent_lastmonth']
Campaign_SG_lastmonth['CPA'] = Campaign_SG_lastmonth['Amount_spent_lastmonth'] / Campaign_SG_lastmonth['Purchases_lastmonth']
Campaign_SG_lastmonth['CR(%)'] = (Campaign_SG_lastmonth['Purchases_lastmonth'] / Campaign_SG_lastmonth['Link_clicks_lastmonth']) * 100
# Change order
Campaign_SG_lastmonth = Campaign_SG_lastmonth.reindex(columns=columns_order_lastmonth)
Campaign_SG_lastmonth = Campaign_SG_lastmonth.reindex(index=index_order)
# Transpose
Campaign_SG_lastmonth = Campaign_SG_lastmonth.transpose()


# Combine two series
# Change the series into same index, in order to concat
Campaign_SG_lastmonth.index = Campaign_SG_thismonth.index
Campaign_SG = pd.concat([Campaign_SG_lastmonth, Campaign_SG_thismonth],axis=1)


# Calculate growth
Growth = (Campaign_SG_thismonth - Campaign_SG_lastmonth) / Campaign_SG_lastmonth * 100
# Combine series
Campaign_SG = pd.concat([Campaign_SG, Growth],axis=1)


# Change index
Campaign_SG = Campaign_SG.reset_index()
# Change columns name
Campaign_name = Campaign_SG.columns
Number = int((Campaign_SG.shape[1] - 1) / 3)
Column_name = ['Campaign_SG'] + [Lastmonth] * Number + [Thismonth] * Number +  ['Growth(%)'] * Number

Campaign_SG = pd.DataFrame([Campaign_name] + Campaign_SG.values.tolist(), columns = Column_name)

# Round 2
Campaign_SG = Campaign_SG.round(2)
```
