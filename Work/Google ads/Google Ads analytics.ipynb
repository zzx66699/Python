```python
import pandas as pd

df = pd.read_excel('raw_data.xlsx')
df

r = df[["Campaign",
        "Cost",
        "Cost (Compare to)",
        "Cost (Change %)",
        "Impr.",
        "Impr. (Compare to)", 
        "Impr. (Change %)",
        "Clicks",
        "Clicks (Compare to)", 
        "Clicks (Change %)",
        "Add to basket",
        "Add to basket (Compare to)",
        "Add to basket (Change %)",
        "Purchase",
        "Purchase (Compare to)",
        "Purchase (Change %)",
        "Revenue",
        "Revenue (Compare to)",
        "Revenue (Change %)"]]

r = r.dropna(subset=['Campaign'])
r = r.reset_index(drop=True)

r
```

```python
sum_row = pd.DataFrame(columns=r.columns)
sum_row.loc[0, 'Campaign'] = 'Sum'
for column in r.columns:
    if column != 'Campaign':
        sum_row.loc[0, column] = r[column].sum()
sum_row["Cost (Change %)"] = sum_row["Cost"] / sum_row["Cost (Compare to)"] - 1
sum_row["Clicks (Change %)"] = sum_row["Clicks"] / sum_row["Clicks (Compare to)"] - 1
sum_row["Impr. (Change %)"] = sum_row["Impr."] / sum_row["Impr. (Compare to)"] - 1
sum_row["Add to basket (Change %)"] = sum_row["Add to basket"] / sum_row["Add to basket (Compare to)"] - 1
sum_row["Purchase (Change %)"] = sum_row["Purchase"] / sum_row["Purchase (Compare to)"] - 1
sum_row["Revenue (Change %)"] = sum_row["Revenue"] / sum_row["Revenue (Compare to)"] - 1
```

```python
r = pd.concat([r, sum_row], ignore_index=True)
```

```python
sg = r[r['Campaign'].str.contains('SG')]
sum_row = pd.DataFrame(columns=sg.columns)
sum_row.loc[0, 'Campaign'] = 'Sum'
for column in sg.columns:
    if column != 'Campaign':
        sum_row.loc[0, column] = sg[column].sum()
sum_row["Cost (Change %)"] = sum_row["Cost"] / sum_row["Cost (Compare to)"] - 1
sum_row["Clicks (Change %)"] = sum_row["Clicks"] / sum_row["Clicks (Compare to)"] - 1
sum_row["Impr. (Change %)"] = sum_row["Impr."] / sum_row["Impr. (Compare to)"] - 1
sum_row["Add to basket (Change %)"] = sum_row["Add to basket"] / sum_row["Add to basket (Compare to)"] - 1
sum_row["Purchase (Change %)"] = sum_row["Purchase"] / sum_row["Purchase (Compare to)"] - 1
sum_row["Revenue (Change %)"] = sum_row["Revenue"] / sum_row["Revenue (Compare to)"] - 1
sg = pd.concat([sg, sum_row], ignore_index=True)
sg
```

```python
output = [r, sg, uk]
writer = pd.ExcelWriter('Google.xlsx', engine='xlsxwriter')
sheet_name = 'Sheet1'
start_row = 0
for a in output:
    a.to_excel(writer, sheet_name=sheet_name, index=False, startrow=start_row)
    start_row += a.shape[0] + 2
writer.save()
```
